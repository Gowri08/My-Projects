```python
import cv2
import matplotlib.pyplot as plt
```

```python
dir=r"C:\Users\gowri\Downloads\archive (3)\training_set\training_set"
```

```python
category=["cats","dogs"]
df=[]
```

```python
for i in category:
    path=os.path.join(dir,i)
    for img in os.listdir(path):
        try:
            img_path=os.path.join(path,img)
            label=category.index(i)
            img_arr=cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
            img_arr=cv2.resize(img_arr,(60,60))
            df.append([img_arr,label])
        except Exception as e:
            print(img_path)       
```

```python
import random
random.shuffle(df)
```

```python
x,y=[],[]
for i,j in df:
    x.append(i)
    y.append(j)
```

```python
x
```

```python
import numpy as np
y=np.array(y)
y.reshape(-1,1)
```

```python
x=tf.keras.utils.normalize(x)
x=x.reshape(-1,60,60,1)
x
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Dense,Flatten
```

```python
model=Sequential([
    Conv2D(64,(3,3),activation="relu"),
    MaxPooling2D((2,2)),
    Conv2D(64,(3,3),activation="relu"),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(128,activation="relu"),
    Dense(2,activation="sigmoid")
])
```

```python
model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=10,batch_size=250)
```

```python
def img(path):
    img_output=cv2.imread(path,cv2.IMREAD_GRAYSCALE)
    img_output=cv2.resize(img_output,(60,60))
    img_output=np.array(img_output).reshape(-1,60,60,1)
    return img_output
```

```python
def prediction(path):
    pred=model.predict(img(path))
    print(category[pred.argmax()])
```

```python
model.save("dog_vs_cat.model")
```
